```python
from sklearn.svm import SVR
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generating linear data with Gaussian noise
# data points should be around the line 6x+10

# noise with variance 5
epsilon = np.random.normal(0,np.sqrt(5),size=10)
x = np.linspace(0,9,10)
y = 6*x+10+epsilon

print(y)
```

```python
regressor = SVR(kernel = 'linear')
regressor.fit(x.reshape(-1, 1),y)
```

```python
plt.scatter(x,y, label='True data')
plt.plot(x,regressor.predict(x.reshape(-1,1)), color='red', label='SVR margin')
plt.title("Linear SVR on dummy test data")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

print("Predicted slope: ", regressor.coef_)
print("Score: ", regressor.score(x.reshape(-1,1),y)) # 1 is the best score
```
